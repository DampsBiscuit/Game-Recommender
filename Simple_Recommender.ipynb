```python
import pandas as pd
from tabulate import tabulate

metadata = pd.read_csv('steam_games.csv', low_memory=True)

metadata['original_price'] = metadata['original_price'].replace(['Free', 'Not Available', 'Free to Play', 'Demo'], '0')

metadata['original_price'] = metadata['original_price'].str.replace('$', '').str.strip()

metadata['original_price'] = pd.to_numeric(metadata['original_price'], errors='coerce')

metadata['original_price'] = metadata['original_price'].fillna(metadata['original_price'].median())

mean_price = metadata['original_price'].mean()

print(mean_price)
```

```python
def recommend_games_closest_to_mean():
    price_range = 10
    similar_price_games = metadata[(metadata['original_price'] >= mean_price - price_range) & (metadata['original_price'] <= mean_price + price_range)]
  
    return similar_price_games[['name', 'release_date', 'developer', 'genre', 'original_price']].head(20)

recommended_games = recommend_games_closest_to_mean()

recommended_games = recommended_games.sort_values('original_price', ascending=True)
recommended_games[['name', 'release_date', 'developer', 'genre', 'original_price']].head(20)
```
